```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("HR_comma_sep.csv")
```

```python
df.rename(columns={"average_montly_hours": "average_monthly_hours"}, inplace=True)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["satisfaction_level"], y=df["average_monthly_hours"], hue=df["left"],
                alpha=0.6, palette={0: "blue", 1: "red"})
plt.xlabel("Satisfaction Level")
plt.ylabel("Average Monthly Hours")
plt.title("Satisfaction Level vs. Average Monthly Hours (Colored by Attrition)")
plt.legend(title="Left", labels=["Stayed", "Left"])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="salary", hue="left", data=df, palette="coolwarm")
plt.xlabel("Salary Level")
plt.ylabel("Count")
plt.title("Employee Attrition by Salary Level")
plt.legend(title="Left", labels=["Stayed", "Left"])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(y="Department", hue="left", data=df, palette="coolwarm")
plt.xlabel("Count")
plt.ylabel("Department")
plt.title("Employee Attrition by Department")
plt.legend(title="Left", labels=["Stayed", "Left"])
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x="promotion_last_5years", hue="left", data=df, palette="coolwarm")
plt.xlabel("Promotion in Last 5 Years")
plt.ylabel("Count")
plt.title("Effect of Promotion on Employee Attrition")
plt.legend(title="Left", labels=["Stayed", "Left"])
plt.show()
```
